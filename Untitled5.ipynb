```python
!pip install pandas numpy matplotlib seaborn openpyxl Faker
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from faker import Faker
from datetime import datetime, timedelta

# Configurações de visualização
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = [12, 6]
```

```python
# --- PARTE 2: SIMULAÇÃO E GERAÇÃO DO DATASET ---

print("### 1. Geração do Dataset Fictício (10.000 Transações) ###")

# Configuração da Simulação
fake = Faker('pt_BR')
num_transacoes = 10000
start_date = datetime(2024, 1, 1)
end_date = datetime(2024, 12, 31)

def random_date(start, end):
    return start + timedelta(seconds=np.random.randint(0, int((end - start).total_seconds())))

data = []
categorias = ['Eletrônicos', 'Vestuário', 'Alimentos', 'Casa e Decoração', 'Beleza']
canais = ['Site Próprio', 'Marketplace A', 'Marketplace B', 'Instagram']
regioes = ['Sudeste', 'Sul', 'Nordeste', 'Centro-Oeste', 'Norte']

for i in range(num_transacoes):
    categoria = np.random.choice(categorias, p=[0.3, 0.25, 0.15, 0.15, 0.15]) # Eletrônicos mais frequentes

    # Simulação de Receita e Custo baseada na Categoria
    if categoria == 'Eletrônicos':
        receita_base = np.random.uniform(500, 3000)
        margem_custo = np.random.uniform(0.55, 0.70) # Margem menor (custo alto)
    elif categoria == 'Vestuário':
        receita_base = np.random.uniform(100, 800)
        margem_custo = np.random.uniform(0.30, 0.50)
    else: # Outros
        receita_base = np.random.uniform(50, 400)
        margem_custo = np.random.uniform(0.20, 0.45)

    quantidade = np.random.randint(1, 5)

    data.append({
        'ID_Transacao': f'TRX-{10000 + i}',
        'Data_Venda': random_date(start_date, end_date),
        'ID_Cliente': f'CLI-{np.random.randint(100, 2000)}',
        'Produto_SKU': f'SKU-{np.random.randint(100, 500)}',
        'Nome_Produto': fake.word().capitalize() + ' Pro',
        'Categoria': categoria,
        'Canal_Venda': np.random.choice(canais, p=[0.4, 0.3, 0.15, 0.15]),
        'Regiao_Cliente': np.random.choice(regioes),
        'Receita_Liquida': round(receita_base / quantidade, 2), # Receita por unidade
        'Custo_Unitario_Produto': round(receita_base * margem_custo / quantidade, 2),
        'Quantidade': quantidade,
        'Custo_Logistico': round(np.random.uniform(10, 50) * (1 if categoria == 'Alimentos' else 1.5), 2),
        'Custo_Marketing_Atribuido': round(np.random.uniform(5, 25), 2),
    })

df = pd.DataFrame(data)
print(f"Dataset Simulado criado com {df.shape[0]} linhas.")
print("\nDataFrame Inicial:")
print(df.head())
```

```python
# --- PARTE 3: ENGENHARIA DE VARIÁVEIS (CÁLCULO DO LUCRO) ---

print("\n### 2. Engenharia de Variáveis (Cálculo do Lucro) ###")

# 3.1 Receita e Custo Totais por Transação
df['Receita_Total'] = df['Receita_Liquida'] * df['Quantidade']
df['Custo_Total_Produto'] = df['Custo_Unitario_Produto'] * df['Quantidade']

# 3.2 Custo Total da Transação
df['Custo_Total_Transacao'] = df['Custo_Total_Produto'] + df['Custo_Logistico'] + df['Custo_Marketing_Atribuido']

# 3.3 Lucro Líquido
df['Lucro_Liquido'] = df['Receita_Total'] - df['Custo_Total_Transacao']

# 3.4 Margem de Lucro Percentual
df['Margem_Lucro_Percentual'] = (df['Lucro_Liquido'] / df['Receita_Total']) * 100

print("\nNovas Colunas Calculadas:")
print(df[['Receita_Total', 'Lucro_Liquido', 'Margem_Lucro_Percentual']].head())

# Formatação da data para análise temporal
df['Mes_Ano'] = df['Data_Venda'].dt.to_period('M')
```

```python
# --- PARTE 4: ANÁLISE EXPLORATÓRIA DE DADOS (EDA) ---

print("\n### 3. Análise Exploratória de Dados (EDA) ###")

# 4.1 Análise Temporal do Lucro
lucro_mensal = df.groupby('Mes_Ano')['Lucro_Liquido'].sum().reset_index()
lucro_mensal['Mes_Ano'] = lucro_mensal['Mes_Ano'].astype(str) # Conversão para string para plotagem

plt.figure(figsize=(14, 6))
sns.lineplot(x='Mes_Ano', y='Lucro_Liquido', data=lucro_mensal, marker='o')
plt.title('Tendência Mensal do Lucro Líquido')
plt.xlabel('Mês')
plt.ylabel('Lucro Líquido (R$)')
plt.xticks(rotation=45)
plt.show()

# 4.2 Lucratividade por Categoria (Gráfico de Barras)
lucro_categoria = df.groupby('Categoria').agg(
    Lucro_Total=('Lucro_Liquido', 'sum'),
    Margem_Media=('Margem_Lucro_Percentual', 'mean')
).reset_index().sort_values(by='Lucro_Total', ascending=False)

plt.figure(figsize=(12, 5))
sns.barplot(x='Categoria', y='Lucro_Total', data=lucro_categoria, palette='viridis')
plt.title('Lucro Líquido Total por Categoria de Produto')
plt.ylabel('Lucro Líquido (R$)')
plt.show()

# 4.3 Margem de Lucro por Canal de Venda (Visualizando Otimização)
margem_canal = df.groupby('Canal_Venda')['Margem_Lucro_Percentual'].mean().reset_index().sort_values(by='Margem_Lucro_Percentual', ascending=False)

plt.figure(figsize=(10, 5))
sns.barplot(x='Canal_Venda', y='Margem_Lucro_Percentual', data=margem_canal, palette='plasma')
plt.title('Margem de Lucro Percentual Média por Canal de Venda')
plt.ylabel('Margem de Lucro Média (%)')
plt.show()
```
